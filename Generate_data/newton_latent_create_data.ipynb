```python

# Constants
world_width  = 800
world_height = 600
xCenter      = world_width / 2

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((world_width, world_height))
clock  = pygame.time.Clock()

# Create a physics space with initial gravity set to 0
space = pymunk.Space()
space.gravity = (0, 0)  # No gravity initially

# Create the ground
ground = pymunk.Segment(space.static_body, (0, world_height - 10), (world_width, world_height - 10), 1)
ground.elasticity = 0.1
space.add(ground)

# Function to generate random dimensions for blocks
def dim():
    return random.randint(50, 100)

# Function to determine the X position of a block
def xpos(prev_block):
    prev_w = prev_block[0]  # width of the previous block
    prev_x = prev_block[1]  # X position of the previous block
    return random.randint(prev_x - prev_w, prev_x + prev_w)

# Function to determine the Y position of a block
def ypos(prev_block, h):
    prev_y = prev_block[2]  # Y position of the previous block
    prev_h = prev_block[3]  # height of the previous block
    return prev_y - (prev_h)

# Function to add a new block
def add_block(prev_block, is_first):
    w = dim()
    h = dim()
    x = xCenter if is_first else xpos(prev_block)
    y = ypos(prev_block, h)
    return (w, x, y, h)

# Function to create the world with blocks stacked
def make_tower_world():
    block1 = add_block((0, xCenter, world_height - 10, 10), True)  # Ground block
    block2 = add_block(block1, False)
    block3 = add_block(block2, False)
    block4 = add_block(block3, False)
    block5 = add_block(block4, False)
    return [ground, block1, block2, block3, block4, block5]

# Main loop to animate the tower world
def animate_tower():
    blocks = make_tower_world()
    
    # Add blocks as dynamic bodies
    for block in blocks[1:]:
        mass = 1
        moment = pymunk.moment_for_box(mass, (block[0], block[3]))  # Calculate moment for box
        body = pymunk.Body(mass, moment)
        body.position = (block[1], block[2])  # Set position of the body
        body.velocity = (0, 0)  # Ensure no initial velocity
        shape = pymunk.Poly.create_box(body, (block[0], block[3]))  # Create the shape
        shape.elasticity = 0.1  # Set elasticity (bounciness)
        space.add(body, shape)  # Add body and shape to space

    # Run the simulation to display initial blocks
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:  # Press space to activate gravity
                    space.gravity = (0, 50)  # Apply gravity
                    print("Gravity activated!")
        
        screen.fill((255, 255, 255))  # Clear screen
        space.debug_draw(pymunk.pygame_util.DrawOptions(screen))  # Draw the simulation
        pygame.display.flip()

        space.step(1/60.0)  # Simulate physics at 60 FPS
        clock.tick(60)

    pygame.quit()

# Start the simulation
animate_tower()
```

```python

# Constants
world_width  = 800
world_height = 600
xCenter      = world_width / 2

# Initialize Pygame
pygame.init()

# Ensure output directory exists
output_dir = 'videos'
os.makedirs(output_dir, exist_ok=True)

# Create the Pygame screen only once, before the loop
screen = pygame.display.set_mode((world_width, world_height))
clock = pygame.time.Clock()

# Function to generate random dimensions for blocks
def dim():
    return random.uniform(50, 100)

# Function to determine the X position of a block
def xpos(prev_block):
    prev_w = prev_block[0]  # width of the previous block
    prev_x = prev_block[1]  # X position of the previous block
    return random.uniform(prev_x - prev_w, prev_x + prev_w)

# Function to determine the Y position of a block
def ypos(prev_block, h):
    prev_y = prev_block[2]  # Y position of the previous block
    prev_h = prev_block[3]  # height of the previous block
    return prev_y - (prev_h)

# Function to add a new block
def add_block(prev_block, is_first):
    w = dim()
    h = dim()
    x = xCenter if is_first else xpos(prev_block)
    y = ypos(prev_block, h)
    return (w, x, y, h)

# Function to create the world with blocks stacked
def make_tower_world():
    block1 = add_block((0, xCenter, world_height - 10, 10), True)  # Ground block
    block2 = add_block(block1, False)
    block3 = add_block(block2, False)
    block4 = add_block(block3, False)
    block5 = add_block(block4, False)
    return [ground, block1, block2, block3, block4, block5]

# Initialize the video writer (output video file)
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# Loop for 10,000 examples
for i in range(1, 10001):
    # Create video file name
    video_filename = os.path.join(output_dir, f'simulation_{i:05d}.avi')
    out = cv2.VideoWriter(video_filename, fourcc, 30.0, (world_width, world_height))

    # Create a physics space with initial gravity set to 0
    space = pymunk.Space()
    space.gravity = (0, 50)  # Gravity immediately activated

    # Create the ground
    ground = pymunk.Segment(space.static_body, (0, world_height - 10), (world_width, world_height - 10), 1)
    ground.elasticity = 0.9
    space.add(ground)

    blocks = make_tower_world()

    # Add blocks as dynamic bodies
    for block in blocks[1:]:
        mass = 1
        moment = pymunk.moment_for_box(mass, (block[0], block[3]))  # Calculate moment for box
        body = pymunk.Body(mass, moment)
        body.position = (block[1], block[2])  # Set position of the body
        body.velocity = (0, 0)  # Ensure no initial velocity
        shape = pymunk.Poly.create_box(body, (block[0], block[3]))  # Create the shape
        shape.elasticity = 0.9  # Set elasticity (bounciness)
        space.add(body, shape)  # Add body and shape to space

    # Collect data for later disentangling the latent space
    velocities = []
    positions = []
    times = []  # List to store time at each frame

    # Run the simulation for 1200 frames (20 seconds)
    running = True
    frame_count = 0
    start_time = pygame.time.get_ticks()  # Start time at the beginning of the simulation
    
    while running and frame_count < 600:  # Stop after 20 seconds (1200 frames)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Record velocity, position, and time for each block
        for body in space.bodies:
            velocities.append(body.velocity)
            positions.append(body.position)

        # Record the time in seconds
        elapsed_time = (pygame.time.get_ticks() - start_time) / 1000.0  # Time in seconds
        times.append(elapsed_time)

        # Draw the simulation on screen
        screen.fill((255, 255, 255))  # Clear screen
        space.debug_draw(pymunk.pygame_util.DrawOptions(screen))  # Draw the simulation
        pygame.display.flip()

        # Capture the screen as a frame and write to video
        frame = pygame.surfarray.array3d(pygame.display.get_surface())  # Get the frame
        frame = np.transpose(frame, (1, 0, 2))  # Transpose to match OpenCV's BGR format
        out.write(frame)  # Write the frame to the video

        space.step(1/60.0)  # Simulate physics at 60 FPS
        clock.tick(60)

        frame_count += 1

    out.release()  # Release the video writer

    # Optionally, save velocity, position, and time data to files (e.g., CSV)
    np.savetxt(f"{output_dir}/velocities_{i:05d}.csv", np.array(velocities), delimiter=",")
    np.savetxt(f"{output_dir}/positions_{i:05d}.csv", np.array(positions), delimiter=",")
    np.savetxt(f"{output_dir}/times_{i:05d}.csv", np.array(times), delimiter=",")  # Save time data

# Call pygame.quit() after the loop is finished
pygame.quit()
```
